## Robbie the Robot: Counting Paths with Dynamic Programming

Robbie starts in the **upper-left corner** of an **8×8 grid** and wants to reach the **lower-right corner**, where the **Prize** is located.

Robbie may move only:

- **right**
- **down**

This is a classic dynamic programming problem because the number of ways to reach a square depends on the number of ways to reach earlier squares.

### State Definition

Let

\[
dp[r][c]
\]

mean:

> the number of different ways to reach row `r`, column `c`

where:

- `(0,0)` is the start
- `(7,7)` is the Prize

### Recurrence

For any non-edge square:

\[
dp[r][c] = dp[r-1][c] + dp[r][c-1]
\]

because Robbie can arrive only:

- from above
- from the left

### Base Cases

- `dp[0][0] = 1`
- first row = all `1`
- first column = all `1`

### Goal

Compute the number of ways Robbie can reach `(7,7)`.

```python
def robbie_count_paths(rows=8, cols=8):
    """Count paths for Robbie on an empty grid using bottom-up DP."""
    dp = [[0] * cols for _ in range(rows)]

    # Start cell
    dp[0][0] = 1

    # First row
    for c in range(1, cols):
        dp[0][c] = 1

    # First column
    for r in range(1, rows):
        dp[r][0] = 1

    # Fill the rest
    for r in range(1, rows):
        for c in range(1, cols):
            dp[r][c] = dp[r - 1][c] + dp[r][c - 1]

    return dp
```

## Example 1: Robbie on an Empty 8×8 Grid

In an empty grid, Robbie never encounters obstacles.

This produces a table where each cell contains the sum of:

- the cell above
- the cell to the left

This is closely related to Pascal-style addition in two dimensions.

```python
dp = robbie_count_paths(8, 8)

print("DP table:")
for row in dp:
    print(row)

print()
print("Number of paths to the Prize:", dp[7][7])
```

## Expected Result for Example 1

The bottom-right corner should contain:

\[
3432
\]

So Robbie has **3432 different ways** to reach the Prize on an empty 8×8 grid.

## Robbie with Blocked Squares

Now we make the problem more interesting by adding blocked cells.

Robbie still starts at `(0,0)` and wants to reach `(7,7)`, but may **not step on blocked squares**.

Blocked cells use **0-based indexing**.

### Blocked Cells

- `(2,1)`
- `(5,4)`

### Modified Recurrence

If `(r,c)` is blocked, then:

\[
dp[r][c] = 0
\]

Otherwise:

\[
dp[r][c] = dp[r-1][c] + dp[r][c-1]
\]

### Important Idea

Blocked cells act like broken states in the DP table. Their effect propagates forward because later squares may depend on them.

```python
def robbie_count_paths_with_blocks(rows=8, cols=8, blocked=None):
    """Count paths for Robbie on a grid with blocked squares."""
    if blocked is None:
        blocked = set()
    else:
        blocked = set(blocked)

    dp = [[0] * cols for _ in range(rows)]

    # If the start is blocked, no path exists
    if (0, 0) in blocked:
        return dp

    dp[0][0] = 1

    # First row
    for c in range(1, cols):
        if (0, c) not in blocked:
            dp[0][c] = dp[0][c - 1]

    # First column
    for r in range(1, rows):
        if (r, 0) not in blocked:
            dp[r][0] = dp[r - 1][0]

    # Fill the rest
    for r in range(1, rows):
        for c in range(1, cols):
            if (r, c) in blocked:
                dp[r][c] = 0
            else:
                dp[r][c] = dp[r - 1][c] + dp[r][c - 1]

    return dp
```

## Example 2: Robbie with Obstacles

We now place two blocked squares at:

- `(2,1)`
- `(5,4)`

The dynamic programming table will still be filled row by row, but blocked cells will contain `0`.

```python
blocked = {(2, 1), (5, 4)}
dp_blocked = robbie_count_paths_with_blocks(8, 8, blocked)

print("Blocked cells:", blocked)
print()
print("DP table:")

for r, row in enumerate(dp_blocked):
    display_row = []
    for c, value in enumerate(row):
        if (r, c) in blocked:
            display_row.append("X")
        else:
            display_row.append(value)
    print(display_row)

print()
print("Number of valid paths to the Prize:", dp_blocked[7][7])
```

## Expected Result for Example 2

With blocked squares at `(2,1)` and `(5,4)`, the bottom-right corner should contain:

\[
1386
\]

So Robbie has **1386 valid paths** to the Prize while avoiding the obstacles.

## Complexity

For both versions of the Robbie problem:

- **Time complexity:** \(O(rows \cdot cols)\)
- **Space complexity:** \(O(rows \cdot cols)\)

For an 8×8 grid this is small, but the same method scales naturally to larger grids.

## Key Takeaways

- Dynamic programming works by solving smaller subproblems once and reusing them.
- In the Robbie problem, each square depends on previously solved neighboring squares.
- Adding obstacles is easy: blocked states simply become `0`.
- This is a clean bridge from Pascal-style counting to more realistic constrained problems.
